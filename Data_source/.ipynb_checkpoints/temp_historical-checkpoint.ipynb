#### Import Dependencies

```python
import pandas as pd
import os
import csv
import numpy as np
```

#### Store filepath in a variable

```python
csv_path = os.path.join("GlobalLandTemperaturesByCountry - Copy.csv")
csv_path
```

```python
temp_df = pd.read_csv(csv_path)
temp_df
```

```python
temp_df.head()
```

```python
countries_df = temp_df.loc[(temp_df["Country"]== "Brazil") |(temp_df["Country"]== "United States"),:]
countries_df
```

```python
countries_df['dt'] = pd.to_datetime(countries_df['dt'])
countries_df['Year'] = countries_df['dt'].dt.year
countries_df
```

```python
frame = countries_df.reset_index()
frame
```

```python
frame_v1 = frame.drop(frame.index[0:2016])
frame_v1
```

```python
frame_v2 = frame_v1.reset_index()
frame_v2
```

```python
frame_v3 = frame_v2.drop(frame_v2.index[165:2941])
frame_v3
```

```python
frame_temperature_f = frame_v3.drop(columns=['level_0', 'index', 'dt', 'AverageTemperatureUncertainty'])
frame_temperature_f
```

```python
frame_temperature_f2 = frame_temperature_f.groupby(['Country','Year']).max()
frame_temperature_f2
```

```python
frame_temperature_f3 = frame_temperature_f2.rename(columns={'AverageTemperature': 'Max Annual Temperature C'})
frame_temperature_f3
```

```python
frame_temperature_final = frame_temperature_f3.reset_index()
frame_temperature_final
```

```python
output_file = os.path.join("MaxTemperature.csv")
frame_temperature_final.to_csv(output_file, index=False, header=True)
```
