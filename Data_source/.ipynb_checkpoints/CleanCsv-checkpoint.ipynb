```python
import pandas as pd
import os
```

```python
file = os.path.join("Output","emission data.csv")
```

```python
df = pd.read_csv(file)
df
```

```python
df.count()
```

```python
df_years=df[["Country","2001", "2002","2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013"]]
df_years
```

```python
df_countries=df_years.loc[(df["Country"] == "Russia") | (df["Country"] == "Brazil") | (df["Country"] == "Canada") | (df["Country"] == "China") | (df["Country"] == "Australia") | (df["Country"] == "United States"), :]
```

```python
output_file = os.path.join("Output", "CleanEmissionData.csv")

df_countries.to_csv(output_file, index=False, header=True)
```
