```python
import pandas as pd
import os
```

```python
file = os.path.join("emission data.csv")
```

```python
df = pd.read_csv(file)
df
```

```python
# setting column Country as index"
df.set_index("Country", inplace=True)
df
```

```python
# selecting desired columns"
df_yrs=df[["2000","2001", "2002","2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013"]]
df_yrs
```

```python
# selecting desired rows
df_countries=df_yrs.loc[["Brazil","United States"], :]
df_countries
```

```python
# Stacking columns
df_1=df_countries.stack()
df_1
```

```python
df_2=df_1.reset_index()
df_2.columns

df_3=df_2.rename(columns={"level_1":"Year", 0:"Tonnes"})
df_3
```

```python
output_file = os.path.join("CleanEmissionData.csv")

df_3.to_csv(output_file, index=False, header=True)
```
