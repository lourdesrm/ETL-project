```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import pymongo
from splinter import Browser
```

```python
table_fire = pd.read_html('https://www.ncdc.noaa.gov/societal-impacts/wildfires/ytd/0?params[]=acres&params[]=fires&params[]=apf')
table_fire_us = table_fire[0]
```

```python
table_fire_us
```

```python
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200001) & (table_fire_us['Date'] < 200101), 2000)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200101) & (table_fire_us['Date'] < 200201), 2001)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200201) & (table_fire_us['Date'] < 200301), 2002)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200301) & (table_fire_us['Date'] < 200401), 2003)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200401) & (table_fire_us['Date'] < 200501), 2004)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200501) & (table_fire_us['Date'] < 200601), 2005)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200601) & (table_fire_us['Date'] < 200701), 2006)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200701) & (table_fire_us['Date'] < 200801), 2007)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200801) & (table_fire_us['Date'] < 200901), 2008)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 200901) & (table_fire_us['Date'] < 201001), 2009)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 201001) & (table_fire_us['Date'] < 201101), 2010)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 201101) & (table_fire_us['Date'] < 201201), 2011)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 201201) & (table_fire_us['Date'] < 201301), 2012)
table_fire_us['Date'] = table_fire_us['Date'].mask((table_fire_us['Date'] >= 201301), 2013)
```

```python
table_fire_us.head(100)
```

```python
table_fire_us_count = table_fire_us.groupby(['Date']).sum()
```

```python
only_fires = table_fire_us_count.drop(columns=['Acres Burned', 'Acres Burned per Fire'])
```

```python
final_us_count = only_fires.reset_index()
final_us_count
```

```python
final_us_count.to_csv('fire_count_us.csv', index=False)
```
